```python
exclude = np.repeat(False, df.shape[0])
prots = df["leading_razor_protein"]
filter_rev = prots.str.contains("REV*")
filter_rev[pd.isnull(filter_rev)] = False
print(np.sum(filter_rev))
filter_con = prots.str.contains("CON*")
filter_con[pd.isnull(filter_con)] = False
print(np.sum(filter_con))
exclusion_list = [line.rstrip('\n') for line in open("./../pd_exclude.txt")]
pat = reduce((lambda x, y: x + "|" + y), exclusion_list)
blacklist_filter = df["proteins"].str.contains(pat)
blacklist_filter[pd.isnull(blacklist_filter)] = False
print(np.sum(blacklist_filter))
```

```python
#print(np.sum((df["retention_length"] > 5)))
exclude = (filter_rev | filter_con | blacklist_filter | (df["retention_length"] > 5) | (df["pep"] > 0.5))
print(np.sum(exclude))
df["exclude"] = exclude
```

```python
df["exp_id"] = df["raw_file"].map({ind: val for val, ind in enumerate(df["raw_file"].unique())})
df["peptide_id"] = df["sequence"].map({ind: val for val, ind in enumerate(df["sequence"].unique())})

print(len(df["raw_file"].unique()))
print(len(df["sequence"].unique()))
#np.sum(df["peptide_id"].map(df.groupby("peptide_id")["exp_id"].count() < 3))
```

```python
print(df[-(df["exclude"])].shape[0])
#df.groupby(["exclude", "peptide_id"]).get_group(False)
#exps_per_pep = df["peptide_id"].map(df[-(df["exclude"])].groupby("peptide_id")["exp_id"].count())
exps_per_pep = df[-(df["exclude"])].groupby("peptide_id")["exp_id"].unique().apply((lambda x: len(x)))
exps_per_pep = df["peptide_id"].map(exps_per_pep)
print(np.nansum((exps_per_pep < 3)))
exps_per_pep[np.isnan(exps_per_pep)] = 0
print(np.sum((exps_per_pep < 3)))
```
