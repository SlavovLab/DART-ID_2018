```python
from itertools import groupby

def fasta_iter(fasta_name):
    """
    given a fasta file. yield tuples of header, sequence
    """
    fh = open(fasta_name)
    # ditch the boolean (x[0]) and just keep the header or sequence since
    # we know they alternate.
    faiter = (x[1] for x in groupby(fh, lambda line: line[0] == ">"))
    for header in faiter:
        # drop the ">"
        #header = header.next()[1:].strip()
        header = header.__next__()[1:].strip()
        # join all sequence lines to one.
        seq = "".join(s.strip() for s in faiter.__next__())
        yield header, seq
```

```python
fh = open('/gd/Slavov_Lab/Albert/Human_ONLYsp.fasta', 'r')

num_entries = 0
for line in fh:
    if line[0] == '>': num_entries = num_entries + 1
print(num_entries)

fh.close()

headers = np.empty(num_entries, dtype=object)
seqs = np.empty(num_entries, dtype=object)


ff = fasta_iter('/gd/Slavov_Lab/Albert/Human_ONLYsp.fasta')
i = 0
for header, seq in ff:
    headers[i] = header
    seqs[i] = seq
    i=i+1
```

```python
headers = np.array(headers, dtype=object)
prots = [h.split('|')[1] for h in headers]
prots = np.array(prots, dtype=object)
full_prots = [h.split(' ')[0] for h in headers]
full_prots = np.array(full_prots, dtype=object)
descrs = [' '.join(h.split(' ')[1:]) for h in headers]
descrs = np.array(descrs, dtype=object)
```

```python
prev_aa = ''
next_aa = ''

prev_aa_list = np.empty(df.shape[0], dtype=object)
next_aa_list = np.empty(df.shape[0], dtype=object)

for i in range(0, df.shape[0]):
    obs = df.loc[i]
    prots = str(obs['Leading proteins']).split(';')
    #print(prots)
    for prot in prots:
        
        if 'REV' in prot or 'CON' in prot:
            prev_aa = prev_aa + '-'
            next_aa = next_aa + '-'
            continue
        
        ind = db['protein'].str.contains(prot)
        
        if np.sum(ind) == 0:
            prev_aa = prev_aa + '-'
            next_aa = next_aa + '-'
            continue

        seq = db['sequence'].loc[ind].values[0]
        
        m = re.search(r'' + obs['Sequence'] + '', seq)
        
        if m is None:
            prev_aa = prev_aa + '-'
            next_aa = next_aa + '-'
            continue
        
        if m.start() == 0:
            prev_aa = prev_aa + '-'
        else:
            prev_aa = prev_aa + seq[m.start()-1]
        
        if m.end() == len(seq):
            next_aa = next_aa + '-'
        else:
            next_aa = next_aa + seq[m.end()]
    
    prev_aa_list[i] = prev_aa
    next_aa_list[i] = next_aa
    
    prev_aa = ''
    next_aa = ''
        
```

```python
df['prev_aa'] = prev_aa_list
df['next_aa'] = next_aa_list
df
```
